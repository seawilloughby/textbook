```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

###Subplot Examples

```python
total_charts_down = 2
total_charts_across = 3
number_of_charts = 2 * 3
fig = plt.figure()

my_colors = ['red', 'orange', 'yellow', 'green', 'blue', 'indigo', 'violet' ]

for i in xrange(number_of_charts):
    plt.subplot(2, 3, i+1, axisbg=my_colors[i])
    # NEED TO USE i+1 BECAUSE I STARTS AT 0, 
    # AND SUBPLOT LOCATIONS START AT 1.
```

```python
charts_down = 2
charts_across = 4
fig, charts = plt.subplots( charts_down, charts_across )

for i in range(charts_down):
    for j in range(charts_across):
        randomness = np.random.rand(50)
        charts[i,j].hist( randomness )
```

```python

x = np.linspace(0, 5, 10)
y = x ** 2

fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # inset axes

# main figure
axes1.plot(x, y, 'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('title')

# insert
axes2.plot(y, x, 'g')
axes2.set_xlabel('y')
axes2.set_ylabel('x')
axes2.set_title('insert title');
```
